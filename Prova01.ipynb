```python
valor1 = float(input("Digite o primeiro valor: "))
valor2 = float(input("Digite o segundo valor: "))
valor3 = float(input("Digite o terceiro valor: "))

if valor1 >= valor2 and valor2 >= valor3:
    print("A média aritmética dos dois maiores valores é: ", (valor1 + valor2) / 2)
    
elif valor2 >= valor3 and valor3 >= valor1:
    print("A média aritmética dos dois maiores valores é: ", (valor2 + valor3) / 2)

else:
    print("A média aritmética dos dois maiores valores é: ", (valor1 + valor3) / 2)
```
